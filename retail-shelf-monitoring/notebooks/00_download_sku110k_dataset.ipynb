# SKU-110K Dataset Download

This notebook downloads and prepares the **SKU-110K dataset** for retail shelf monitoring.

**Dataset Overview:**
- 11,762 images of retail shelves
- Dense object detection annotations
- Real-world store environments
- Multiple product categories

**Citation:**
Goldman et al., "Precise Detection in Densely Packed Scenes," CVPR 2019

**Source:** https://github.com/eg4000/SKU110K_CVPR19

````python
import os
import requests
import tarfile
import zipfile
from pathlib import Path
from tqdm import tqdm
import shutil

print("✅ Libraries imported successfully")
````

````python
# Dataset Configuration
# Manual download required from:
# https://universe.roboflow.com/new-workspace-okzpb/sku110k/dataset/1

print("📦 SKU-110K Dataset Setup")
print("=" * 70)
print()
print("📥 Download Instructions:")
print("1. Visit: https://universe.roboflow.com/new-workspace-okzpb/sku110k/dataset/1")
print("2. Download SKU110K_fixed.tar.gz (11GB)")
print("3. Place in: data/raw/SKU110K/downloads/")
print("4. Extract: tar -xzf SKU110K_fixed.tar.gz -C data/raw/SKU110K/")
print()
print("=" * 70)

# Directory structure
BASE_DIR = Path("../data/raw/SKU110K")
DOWNLOAD_DIR = BASE_DIR / "downloads"

# Create directories
for directory in [BASE_DIR, DOWNLOAD_DIR]:
    directory.mkdir(parents=True, exist_ok=True)

print(f"\n✅ Base directory: {BASE_DIR}")
print(f"✅ Downloads directory: {DOWNLOAD_DIR}")
````

````python
# Verify the dataset
archive_path = BASE_DIR / "downloads" / "SKU110K_fixed.tar.gz"
extracted_dir = BASE_DIR / "SKU110K_fixed"
images_dir = extracted_dir / "images"

print("🔍 Dataset Verification")
print("=" * 70)

# Check archive
if archive_path.exists():
    file_size_gb = archive_path.stat().st_size / (1024**3)
    print(f"✅ Archive found: {file_size_gb:.2f} GB")
else:
    print(f"❌ Archive not found at: {archive_path}")
    print("   Please download from Roboflow")

# Check extraction
if images_dir.exists():
    print(f"✅ Dataset extracted: {extracted_dir}")
    
    # Count images by split
    train_count = len(list(images_dir.glob("train_*.jpg")))
    val_count = len(list(images_dir.glob("val_*.jpg")))
    test_count = len(list(images_dir.glob("test_*.jpg")))
    total = train_count + val_count + test_count
    
    print()
    print("📊 Image Counts:")
    print(f"   Train: {train_count:,} images")
    print(f"   Val:   {val_count:,} images")
    print(f"   Test:  {test_count:,} images")
    print(f"   Total: {total:,} images")
    
    if total >= 11000:
        print("\n✅ Dataset complete!")
    else:
        print(f"\n⚠️  Expected ~11,762 images, found {total:,}")
else:
    print(f"❌ Dataset not extracted")
    print(f"   Run: tar -xzf {archive_path} -C {BASE_DIR}")

print("=" * 70)
````

````python
from PIL import Image
import matplotlib.pyplot as plt

images_dir = BASE_DIR / "SKU110K_fixed" / "images"

if images_dir.exists():
    # Get sample from each split
    splits = {
        'train': list(images_dir.glob("train_*.jpg"))[:3],
        'val': list(images_dir.glob("val_*.jpg"))[:3],
        'test': list(images_dir.glob("test_*.jpg"))[:3]
    }
    
    fig, axes = plt.subplots(3, 3, figsize=(15, 12))
    fig.suptitle('SKU-110K Dataset Samples', fontsize=16, fontweight='bold')
    
    for row, (split, paths) in enumerate(splits.items()):
        for col, img_path in enumerate(paths[:3]):
            if img_path.exists():
                img = Image.open(img_path)
                axes[row, col].imshow(img)
                axes[row, col].set_title(f"{split}: {img_path.name}", fontsize=10)
                axes[row, col].axis('off')
    
    plt.tight_layout()
    plt.show()
    
    print("✅ Sample images displayed")
else:
    print("⚠️ Images directory not found")
````

## Alternative Datasets (If SKU-110K Unavailable)

If SKU-110K is unavailable from all sources, consider these retail shelf datasets:

### 1. **Grocery Store Dataset** (Recommended Alternative)

- **Size**: 5,125 natural images from grocery stores
- **Classes**: 81 fine-grained product classes
- **Format**: PASCAL VOC annotations
- **Good for**: Product classification + detection

### 2. **WebMarket Dataset**
- **Source**: https://github.com/ParallelDots/Store-Monitoring-System
- **Size**: Retail product images with price tags
- **Good for**: Price verification challenge

### 3. **Freiburg Groceries Dataset**
- **Source**: http://aisdatsets.informatik.uni-freiburg.de/freiburg_groceries_dataset/
- **Size**: 5,000 images, 25 product classes
- **Format**: RGB-D images
- **Good for**: Product recognition

### 4. **Roboflow Retail Collections**
- **Source**: https://universe.roboflow.com/
- Search for: "retail", "shelf", "products", "grocery"
- Many pre-annotated datasets available
